## Conditional Statements & Loops in Python

Conditional statements check conditions using:
- if
- else
- elif

# If Statement

```python
age = 18
if age >= 18:
    print("You can vote")
```

# If-else Statement

```python
age = 16
if age >= 18:
    print("Eligible")
else:
    print("Not Eligible")
```

## Elif Statement

```python
marks = 75
if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

# Nested if

```python
age = 20
citizen = True
if age >= 18:
    if citizen:
        print("Eligible to vote")
```

# Comparison Operators

| Operator | Meaning          |
| -------- | ---------------- |
| `==`     | Equal            |
| `!=`     | Not Equal        |
| `>`      | Greater Than     |
| `<`      | Less Than        |
| `>=`     | Greater or Equal |
| `<=`     | Less or Equal    |


```python
print(10 > 5)
```

## Logical Operators

| Operator | Meaning              |
| -------- | -------------------- |
| `and`    | Both conditions True |
| `or`     | Any one True         |
| `not`    | Reverse condition    |


# Loops

Loops repeat code multiple times.

Types:

for loop
while loop

# for Loop

```python
for i in range(5):
    print(i)
```

# range() Function

```python
range(5)
```

```python
for i in range(1, 6):
    print(i)
```

```python
for i in range(1, 10, 2):
    print(i)
```

## while Loop

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

# Infinite Loop

```python
while True:
    print("Hello")
```

## break Statement

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

## continue Statement

```python
for i in range(1, 6):
    if i == 3:
        continue

    print(i)
```

# pass Statement

```python
for i in range(5):
    pass
```

# Nested Loops
Loop inside another loop.

```python
for i in range(3):
    for j in range(2):
        print(i, j)
```
